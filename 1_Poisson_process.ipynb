## Exponential distribution

First, check if in Julia this distribution is parametrized by its intensity λ (and the mean is 1/λ) or by its scale paramater β (and the mean is β) [[see wikipedia](https://en.wikipedia.org/wiki/Exponential_distribution)].

```julia
fieldnames(Exponential) # names of parameters of the distribution
```

```julia
?Exponential
```

```julia
pgfplots()
plot()
for λ in [0.5,1,1.5]
    plot!(Exponential(1/λ),label="\$\\lambda=\$ $λ")
end
ylims!((0,1.5))
xlims!((0,5))
```

```julia
savefig("graphics_output/1_exponential.pdf") # save the fig in PDF
```

```julia
plot()
for λ in [0.5,1,1.5]
    plot!(Exponential(1/λ),fill=(0, 0.2),label="λ=$λ")
end
ylims!((0,1.5))
xlims!((0,5))
#annotate!(2, 1., L"$t\to\lambda\times e^{-\lambda\,t}$")
```
