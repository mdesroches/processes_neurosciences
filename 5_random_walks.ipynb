### Drunkard's walk

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def RandomWalk(N,d):
    walk = np.cumsum(2*np.random.binomial(1,.5,N)-1)
    return walk
```

```python
plt.plot(np.arange(500),RandomWalk(500,1))
plt.show()
```

```python
endpoints = []
for k in range(3000):
    particularWalk = RandomWalk(500,1)
    endpoints.append(particularWalk[-1])
    plt.plot(np.arange(500),particularWalk,alpha=0.01,color="blue")

plt.grid(True)
plt.show()
```

```python
plt.hist(endpoints, density=True, bins=20)
plt.ylabel('Probability');
```
